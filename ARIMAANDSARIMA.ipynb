##ARIMA AND SARIMA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Sales data
sales = np.array([11, 22.5, 23.6, 27.2, 44.9, 30, 25, 40, 55, 50])

# Create a pandas Series from the sales data
sales_series = pd.Series(sales)

# Fit ARIMA model
arima_model = ARIMA(sales_series, order=(1, 0, 0))
print("arima model")
print(arima_model)
arima_model_fit = arima_model.fit()
print("arima model fit")

print(arima_model_fit)

# Make ARIMA forecast

arima_forecast = arima_model_fit.predict(start=len(sales), end=len(sales)+2)
print("forecast")
print(arima_forecast)

# Fit SARIMA model
sarima_model = ARIMA(sales_series, order=(1, 0, 0))
sarima_model_fit = arima_model.fit()

# Make SARIMA forecast
sarima_forecast = sarima_model_fit.predict(start=len(sales), end=len(sales)+2)

# Print ARIMA forecast
print("ARIMA Forecast:")
print(arima_forecast)

# Print SARIMA forecast

print("sARIMA Forecast:")
print(sarima_forecast)

# Plot the actual sales and forecasts
plt.plot(sales_series, label='Actual Sales')
plt.plot(arima_forecast, label='ARIMA Forecast')
plt.plot(sarima_forecast, label='SARIMA Forecast')

plt.legend()
plt.show()
```

##  brief answers to the ARIMA and SARIMA interview questions:

1. What is ARIMA model and how does it work?
   - ARIMA (AutoRegressive Integrated Moving Average) is a popular time series forecasting model. It combines autoregressive (AR), differencing (I), and moving average (MA) components to capture the temporal dependencies and patterns in the data. ARIMA models aim to predict future values based on the past values and the errors or residuals of the model.

2. Explain the components of ARIMA model.
   - ARIMA model consists of three components:
     - Autoregressive (AR): Represents the relationship between an observation and a fixed number of lagged observations.
     - Integrated (I): Involves differencing the series to make it stationary, removing trends and seasonality.
     - Moving Average (MA): Utilizes the dependency between an observation and a residual error from a moving average model applied to lagged observations.

3. What is the difference between ARIMA and SARIMA models?
   - The main difference lies in the inclusion of a seasonal component. ARIMA models are suitable for non-seasonal time series data, while SARIMA (Seasonal ARIMA) models are designed for time series data with seasonal patterns. SARIMA models include additional seasonal AR, seasonal differencing, and seasonal MA terms to handle the seasonal component.

4. How do you determine the order of an ARIMA model?
   - The order of an ARIMA model is determined by analyzing the autocorrelation function (ACF) and partial autocorrelation function (PACF) plots of the time series data. The order is typically represented as (p, d, q), where p is the order of the AR component, d is the degree of differencing, and q is the order of the MA component.

5. What are the steps involved in fitting an ARIMA model to time series data?
   - The steps include:
     1. Data preparation: Ensure the data is stationary, or perform differencing if needed.
     2. Order determination: Examine ACF and PACF plots to determine the AR, differencing, and MA orders.
     3. Model fitting: Use the determined order to fit the ARIMA model to the data.
     4. Residual analysis: Evaluate the residuals for randomness and autocorrelation.
     5. Forecasting: Use the fitted model to make future predictions.

6. How do you interpret the parameters of an ARIMA model?
   - The parameters of an ARIMA model represent the weights or coefficients assigned to the lagged observations or residuals. The AR parameters indicate the influence of past values on the current value, the MA parameters represent the effect of past errors on the current value, and the differencing parameter represents the extent of differencing performed on the data.

7. What is the purpose of the seasonal component in SARIMA model?
   - The seasonal component in SARIMA models captures the seasonal patterns and fluctuations in the time series data. It accounts for the periodicity and trends that occur at regular intervals within a year or other fixed time frame.

8. How do you determine the seasonal order of a SARIMA model?
   - Similar to determining the order of an ARIMA model, the seasonal order of a SARIMA model is determined by analyzing the seasonal autocorrelation and partial autocorrelation plots. These plots help identify the seasonal AR, seasonal differencing, and seasonal MA orders.
